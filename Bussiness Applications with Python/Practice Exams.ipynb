### Problem 1: Stock Change
Write a function called stock_change() that accepts a list of stock values and determines if the average of the first half of the stocks is less than, or greater than, or equal to the average of the second half of the stock values. Assume that you will always have an even number of stocks in the list. Use print statements to communicate the results.

"average stock value has decreased"    
"average stock value has increased"   
"average stock value is unchanged"    

In addition to the print statements, **return** a list containing the average of the lower half and the average of the upper half rounded to one decimal place.

````python
# Problem 1
def stock_change(data):
  # Find first and second half
  m = len(data) // 2
  first = data[:m]
  second = data[m:]

  # Calculate average of first and second half
  avg_first = sum(first) / len(first)
  avg_second = sum(second) / len(second)

  # Compare and print
  if avg_first < avg_second:
    print("average stock value has increased")
  elif avg_first > avg_second:
    print("average stock value has decreased")
  else:
    print("average stock value is unchanged")

  return [round(avg_first, 1), round(avg_second, 1)]
````

````python
# Test 1
stocks_dn=[4,3,6,4,1,2,3,2]
print(stock_change(stocks_dn))
# prints 'average stock value has decreased
#         [4.2, 2.0]'

# Test 2
stocks_up = [2, 3, 2, 4, 3, 4]
print(stock_change(stocks_up))
# prints 'average stock value has increased
#         [2.3, 3.7]'

# Test 3
stocks_e = [2, 2, 2, 2, 2, 2, 2, 2, 2, 2]
print(stock_change(stocks_e) )
# prints 'average stock value has unchanged
#         [2.0, 2.0]'
````

### Example 2: Data manipulations
Write a function that accepts a dictionary of the following form and returns the overal maximum value along with its data type.
```
data={
    "A":[1,'2','3',4],
    "B":['5',6,7,8]
}
```
Output:
```
find_max_val(data)
(8, int)
```

````python
def find_max_val(dct):
    max_val = float("-inf")
    max_type = None

    for key in dct:
        for val in dct[key]:
            int_val = int(val)
            if int_val > max_val:
                max_val = int_val
                max_type = type(val)

    return max_val, max_type



data={
    "A":[1,'2','3',4],
    "B":['5',6,7,8]
}
find_max_val(data)
# (8, int)
````

Write a function called `extract()` to pull specific information from an XML structure like the `data_string` defined below and return a Python list of tuples of the following form.



````python
data_string = """
<Bookstore>
   <Book ISBN="ISBN-13:978-1599620787" Price="15.23" Weight="1.5">
      <Title>New York Deco</Title>
      <Authors>
         <Author Residence="New York City">
            <First_Name>Richard</First_Name>
            <Last_Name>Berenholtz</Last_Name>
         </Author>
      </Authors>
   </Book>
   <Book ISBN="ISBN-13:978-1579128562" Price="15.80">
      <Remark>
      Five Hundred Buildings of New York and over one million other books are available for Amazon Kindle.
      </Remark>
      <Title>Five Hundred Buildings of New York</Title>
      <Authors>
         <Author Residence="Beijing">
            <First_Name>Bill</First_Name>
            <Last_Name>Harris</Last_Name>
         </Author>
         <Author Residence="New York City">
            <First_Name>Jorg</First_Name>
            <Last_Name>Brockmann</Last_Name>
         </Author>
      </Authors>
   </Book>
</Bookstore>
"""
````

### Example 3: Calculating product sales
This example will be helpful for understanding a portion of problem 3. Further operations are needed to complete the full solution to this problem.

Write a function that will accept a list of the form.
```
results=[('A', 2.5, 10), ('B', 6.4, 20), ('C', 1.3, 15)]
```
Call the function
`sales_calc(results)`

The output should be a dictionary containing the product names as keys and the total sales as values.

```
{'A': 25.0, 'B': 128.0, 'C': 19.5}
```

````python
def sales_calc(lst):
    dct = {}
    for i in lst:
        key = i[0]
        value = i[1]*i[2]
        dct[key] = value
    return dct

results = [("A", 2.5, 10), ("B", 6.4, 20), ("C", 1.3, 15)]
sales_calc(results)
# {'A': 25.0, 'B': 128.0, 'C': 19.5}
````

### Example 4: Finding the extremes
This exercise is similar to the last portion of problem 3 on exam 2. Additional operations are needed to achieve a complete solution.

The objective here is to write a function that will take an object of the following form:
```
data={
    'a':16,
    'b':24,
    'c':10,
    'd':4,
    'e':15,
    'f':3
}
```
Call the function
```
find_extremes(data)
```

The output should be a dictionary with keys labeled `minimum` and `maximum` each of which is associated with a key value pair where the keys are the name of the products with the lowest total sales and the highest total sales and the values are the associated minimum and maximum values.
```
{'minimum': {'f': 3}, 'maximum': {'b': 24}}
```

````python
def find_extremes(dct):
    extremes={}
    import numpy as np
    min = np.inf
    max = -np.inf
    for i in dct:
        if dct[i] < min:
            min = dct[i]
            extremes['minimum'] = {i: min}
        if dct[i] > max:
            max = dct[i]
            extremes['maximum'] = {i: max}
    return extremes



data = {"a": 16, "b": 24, "c": 10, "d": 4, "e": 15, "f": 3}
find_extremes(data)
# {'minimum': {'f': 3}, 'maximum': {'b': 24}}
````

````python
def product_sales(): # the function does not require an argument
    import numpy as np
    import pandas as pd
    import seaborn as sns

    #a. run the query to combine the product name, price, and order quantity into one table
    c.execute("""
    SELECT products.ProductName, products.Price, order_details.Quantity
    FROM order_details
    INNER JOIN products ON order_details.ProductID = products.ProductID
    GROUP BY products.ProductName
    """)

    # store the results of the query as a variable called 'results'
    results = c.fetchall()

    # print the first six records returned by the query
    print(results[:6])

    # print the number of records returned
    print("The total number of records is:", len(results),'\n')

    # b. create a dictionary with the product names as the keys and the total sales as the values (price*quantity)
    sd={}
    for i in results:
        # key=i[--] # keys should be the product names
        key = i[0]
        val = i[1]*i[2] # values are the the price times the quantity for each product
        sd[key] = val

    # print the dictionary
    print(sd,'\n')

    # c. create a dictionary to collect the minimum product sales and the maximum product sales.
    extremes={}

    mn=np.inf # high starting value for the minimum
    mx=0 # low starting value for the maximum

    for i in sd:
        # check if each value in sd is less than the minimum
        # if sd[--]<mn:
        if sd[i]<mn:
            mn= sd[i]
            extremes['minimum']={i:mn} # key is the name of the lowest selling product, value is the total sales of this product
        if sd[i]>mx:
            mx= sd[i]
            extremes['maximum']={i:mx} # key is the name of the highest selling product, value is the total sales of this product
    print(extremes,'\n')

    # provided: run the code to generate a histogram of the values
    sns.histplot(data=list(sd.values()))

product_sales()
````

#### Bar Graph

Grade by `major` stacked by `subject`.
- Use the `sum` function to aggregate the `grade` values.  
- Use the `fillna(0)` method to store missing values as 0s. 

````python
import pandas as pd
import matplotlib.pyplot as plt

def bar_graph(grades_df, colormap=None, return_summary=False):

    # create a pivot table with missing values filled with 0, use .fillna(0)
    stacked_data=pd.pivot_table(
        grades_df,
        values='grade',
        index='major',
        columns='subject',
        aggfunc='sum'
    ).fillna(0)

    # Create the bar plot
    stacked_data.plot(
        kind='bar',
        stacked=True,
        figsize=(8,5),
        colormap=colormap,
        edgecolor='black'
    )

    # Add labels, title, and legend
    plt.xlabel("Major")
    plt.ylabel("Total Grades")
    plt.title("Total Grades by Major (Stacked by Subject)")
    plt.legend(title="Subject", loc='upper left', bbox_to_anchor=(1, 1))
    plt.tight_layout()
    plt.show()

    # Keep this code for autotesting
    if return_summary:
        return stacked_data.shape, stacked_data.values.sum()

    return stacked_data
````

#### 2.b. Perform an Inner Join

   - Create a new table `sales_with_customers` by joining `sales` and `customers` on `customer_id`. Include:  
     - `transaction_id`, `name`, `region`, `product_id`, `quantity`, `total_amount`.
   - Change the `sales_with_customers` table into a Pandas data frame.

````python
def sc_table():
    cursor.execute("""
    SELECT
        s.transaction_id,
        c.name,
        c.region,
        s.product_id,
        s.quantity,
        s.total_amount
    FROM
        sales s
    INNER JOIN
        customers c
    ON
        s.customer_id = c.customer_id
    """)

    sales_with_customers = cursor.fetchall()

    # Get column names from cursor description
    col_names = ["transaction_id", "name", "region", "product_id", "quantity", "total_amount"]

    # Create a pandas DataFrame
    df = pd.DataFrame(sales_with_customers, columns=col_names)
    return df

swc_df = sc_table()
swc_df.head()
````

#### EDA and Preprocessing

Create a function to perform the following analytics tasks.

3. **Preprocess Categorical Variables**:
   - Use `pandas.get_dummies()` to encode the `subject` column as dummy variables. Drop the first column to avoid multicollinearity.

````python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

def eda_and_preprocess(df):
    # Quantitative variables: 'grade' and 'hours_studied'
    df_quant = df[['grade', 'hours_studied', 'absences']]

    # Pairplot
    sns.pairplot(df_quant, corner=True)
    plt.show()

    # Correlation Heatmap
    corr = df_quant.corr()
    mask = np.triu(np.ones_like(corr, dtype=bool))
    plt.figure(figsize=(6, 4))
    sns.heatmap(corr, mask=mask, annot=True, cmap="coolwarm", fmt=".2f", square=True)
    plt.title("Correlation Heatmap")
    plt.show()

    # Preprocess categorical variables
    df = pd.get_dummies(df, columns=['subject'], drop_first=True)
    df['extracurricular_participation'] = df['extracurricular_participation'].map({'yes': 1, 'no': 0})

    return df


# Perform EDA and Visualization
df_processed = eda_and_preprocess(st_df)
df_processed.head()
````

#### Statistical Modeling
The goal of this problem is to implement and test a linear regression model to predict a target variable using specified feature variables. Follow the steps below to complete the task:

**Tasks:**

1. **Understand the Function Inputs**:
   - `df`: The input DataFrame containing the data.
   - `feature_vars`: A list of column names that represent the independent variables (predictors).
   - `target_var`: The name of the dependent variable (target) column.

2. **Construct the Regression Formula**:
   - Combine the feature variables into a string, with each variable separated by `+`.
   - Create the regression formula in the format:  
     ```
     target_var ~ feature_var1 + feature_var2 + ...
     ```
   - Example: `grade ~ hours_studied + extracurricular_participation + absences`

3. **Fit the Linear Model**:
   - Use the `statsmodels.formula.api.ols()` method to fit an ordinary least squares (OLS) regression model using the constructed formula and the input DataFrame.
   - Call the `.fit()` method to generate the model.

4. **Output the Model**:
   - Print the line: `print("Regression Model Summary:")`
   - Return the linear regression model object.

5. **Test with Different Feature and Target Variables**:
   - Ensure your function works with any combination of valid features and target variables from the DataFrame.


````python
import statsmodels.formula.api as smf

def linear_regression_model(df, feature_vars, target_var):
    # Combine the feature variables into a single string separated by ' + '
    features_combined = ' + '.join(feature_vars)

    # Construct the formula for the regression
    formula = target_var + " ~ " + features_combined

    # Fit the linear regression model
    model = smf.ols(formula=formula, data=df).fit()

    # Print and return the model summary
    print("Regression Model Summary:")
    return model
````

````python
# Example Regression
feature_vars = ['hours_studied', 'extracurricular_participation', 'subject_Math', 'subject_Science', 'absences']
target_var = 'grade'

results = linear_regression_model(df_processed, feature_vars, target_var)

# Extract Adjusted R-squared and P-value for the F-statistic
adjusted_r_squared = round(results.rsquared_adj, 2)
f_statistic_p_value = round(results.f_pvalue, 2)

# Print Adjusted R-squared and F-statistic P-value
print(f"\nModel Summary:")
print(f"Adjusted R-squared: {adjusted_r_squared}")
print(f"P-value for F-statistic: {f_statistic_p_value}\n")

# Extract coefficients and p-values
coefficients = results.params
p_values = results.pvalues

# Prepare data for the table
table_data = []
for name, coef, pval in zip(coefficients.index, coefficients.values, p_values.values):
    table_data.append([name, round(coef, 2), round(pval, 2)])


# Formatted Selected Regression Output
from tabulate import tabulate

# Print coefficients and p-values as a table
headers = ["Variable", "Coefficient", "P-value"]
print(tabulate(table_data, headers=headers, tablefmt="grid"))
````

The goal of this problem is to implement and test a linear regression model to predict a target variable using specified feature variables. Follow the steps below to complete the task:


1. **Understand the Function Inputs**:
   - `df`: The input DataFrame containing the data.
   - `feature_vars`: A list of column names that represent the independent variables (predictors).

2. **Construct the Regression Formula**:
   - Combine the feature variables into a string, with each variable separated by `+`.
   - Create the regression formula in the format:  
     ```
     target_var ~ feature_var1 + feature_var2 + ...
     ```
   - Example: `sales_revenue ~ discount + product_rating + source`

3. **Fit the Linear Model**:

````python
import statsmodels.formula.api as smf

def linear_regression_model(df, feature_vars, target_var):
    # Construct the regression formula
    formula = f"{target_var} ~ {' + '.join(feature_vars)}"

    # Fit the linear model
    model = smf.ols(formula=formula, data=df).fit()

    # Print the regression model summary
    print("Regression Model Summary:")
    # print(model.summary())

    # Return the fitted model
    return model
````
