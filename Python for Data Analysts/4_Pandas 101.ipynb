## Pandas Basics

### NumPy Arrays vs Pandas Series

```python
import numpy as np
import pandas as pd
```

```python
my_list = [10,25,35]
x = np.array(my_list)
print(type(x))
print(x, '\n')

y = pd.Series(my_list)
print(type(y))
print(y)
```

```python
index_values = ['Best','Second Best','Third Best']
my_list = [10,25,35]
one = pd.Series(data = my_list,index = index_values)
print(one, '\n')

index_values2 = ['Best','Second Best','Forth Best']
my_list2 = [10,25,40]

two = pd.Series(data = my_list2,index = index_values2)
print(two, '\n')

total = one + two
print(total, '\n')

print(total['Best'])
# total[0] -> deprecated
print(total.iloc[0])
```

```python
df6.groupby("job")["age"].describe()
df6.groupby("job").describe()
```

```python
df6.groupby("education")["outstanding_balance"].sum()
df6.groupby("education")["outstanding_balance"].min()
df6.groupby("education")["outstanding_balance"].max()
df6.groupby("education")["outstanding_balance"].mean()
df6.groupby("education")["outstanding_balance"].median()
df6.groupby("education")["outstanding_balance"].count().sort_values(ascending=True)
```

```python
output = df6.groupby("education").agg({"outstanding_balance": ["sum", "count"]})
output["average"] = (
    output["outstanding_balance"]["sum"] / output["outstanding_balance"]["count"])
output.sort_values(by=("average"), ascending=False)
```

```python
output = df6.groupby("education").agg(
    sum_balance=("outstanding_balance", "sum"),
    count_balance=("outstanding_balance", "count"),
)
output["average"] = (output["sum_balance"] / output["count_balance"])
output.sort_values(by=("average"), ascending=False)

# df6.groupby("education").agg(
#     sum_balance=("outstanding_balance", "sum"),
#     count_balance=("outstanding_balance", "count"),
# ).assign(average=lambda x: (x["sum_balance"] / x["count_balance"]) * 100)
```
