### Problem 6

**6.b.** Determine the minimum number of orders.  
**6.c.** Determine the corresponding customers with the minimum number of orders.  

````python
def customers_orders():
# 6.a. Inner join to gather the customer names along with the count of their orders.
    c.execute("""
    SELECT CustomerName, COUNT(OrderID) AS OrderCount
    FROM Orders o
    INNER JOIN Customers c
    ON o.CustomerID = c.CustomerID
    GROUP BY CustomerName
    """)

    c_o = c.fetchall()
    print(c_o[:5], "\n")

# 6.b. Figure out the minimum number of orders
    from math import inf

    m = inf
    for i in c_o:
        # recall that the number of orders is the value at index 1 in each of the tuples
        if i[1] < m:
            m = i[1]
    print("The minimum number of orders is -->", m, "\n")

# 6.c. Determine the customers with the fewest orders
    for i in c_o:
        # the element at index 0 in each of the tuples is the name of the customer
        if i[1] == m:
            print(i[0])


customers_orders()
````

### Exercise: Probability of Greater

Given a list of numbers and a value, n, write a function that returns the probability of choosing a number greater than or equal to n from the list.  
The probability should be expressed as a percentage, rounded to one decimal place.

> Precent probability of event = (num of favourable outcomes) / (total outcomes) * 100

For example if we have a collection of values [1,8,3,4,19,40] and a number 17,   
the probability that a randomly selected number in the list is greater than 17 is 2/6 or 33.33%.  
There are 2 values greater than 17 in the list and 6 total values.

````python
def probability(lst, num):
    # Calculate the probability
    # return round((len([i for i in lst if i >= num]) / len(lst)) * 100, 1)

    # Count the number of elements greater than or equal to n
    over_count = 0
    for i in lst:
        if i >= num:
            over_count += 1

    # Calculate the probability as a percentage
    total_elements = len(lst)
    prob = (over_count / total_elements) * 100

    # Return the result rounded to one decimal place
    return round(prob, 1)



# Tests
p1=probability([5, 1, 8, 9], 6) #  50.0
print(p1)

p2=probability([7, 4, 17, 14, 12, 3], 16) #  16.7
print(p2)

p3=probability([4, 6, 2, 9, 15, 18, 8, 2, 10, 8], 6) #  70.0
print(p3)
````
